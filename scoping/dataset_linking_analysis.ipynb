## Load Data and Dependencies

```python
#import dependecies
import pandas as pd
import numpy as np
from pathlib import Path

data_folder = Path("../data/")
```

### Address Points (MAR)

```python
#load MAR address points
fileaddresses = data_folder / "address_points/Address_Points.csv"
data_addresses = pd.read_csv(fileaddresses)

print('Columns: ', data_addresses.columns.values)

data_addresses.head()
```

### DDOT Street Functional Classification

```python
#load street functional classification
filestreets = data_folder / "street_functional_classification/DDOT_Street_Functional_Classification.csv"
data_streets = pd.read_csv(filestreets)

print('Columns: ', data_streets.columns.values)

data_streets.head()
```

```python
#is there a streetsegid for every ROADWAYSEGID in data_addresses?
data_addresses_merged = data_addresses.merge(data_streets, how='left', on='ROADWAYSEGID') 

print('Percent entries in the addresses table with no STREETSEGID \
in the street functional classification table: ', data_addresses_merged['STREETSEGID'].isna().sum()/len(data_addresses_merged['STREETSEGID'])*100)
```

### 311 Data

```python
#load 311 data
file311 = data_folder / "311_requests/311_City_Service_Requests_in_2020.csv"
data_311 = pd.read_csv(file311)

print('Columns: ', data_311.columns.values)

data_311.head()
```

```python
#Print some top-level information about some of the data of interest
#how often is the MARADDRESSREPOSITORYID popoulated in the 311 data?
print('Number of rows where MARADDRESSREPOSITORYID field is nan: ', np.sum(data_311['MARADDRESSREPOSITORYID'].isna()))
```

```python
#checking that MAR ID's all exist in the address points database
data_311_merged = data_311.merge(data_addresses, how='left', left_on='MARADDRESSREPOSITORYID', right_on='ADDRESS_ID')

#quick sanity check that the addresses in the large look the same (i.e. we have the right id's in the join)
data_311_merged[['FULLADDRESS', 'STREETADDRESS']].head()
```

```python
#checking that MAR ID's all exist in the address points database
data_311_merged = data_311.merge(data_addresses, how='left', left_on='MARADDRESSREPOSITORYID', right_on='ADDRESS_ID')

print('Number of entries in 311 data that did not map to a ROADWAYSEGID in the MAR addresses table: ', pd.isnull(data_311_merged['ROADWAYSEGID']).sum())
print('Percent of entries in 311 data that did not map to a ROADWAYSEGID in the MAR addresses table: ', \
      pd.isnull(data_311_merged['ROADWAYSEGID']).sum()/len(data_311_merged['ROADWAYSEGID'])*100)

#looking at entries where the tables didn't join
merge_failed = data_311_merged.loc[data_311_merged['ROADWAYSEGID'].isna()==True]
merge_failed[['MARADDRESSREPOSITORYID', 'STREETADDRESS']].head()
```

```python
data_311_merged = data_311_merged.merge(data_streets, how='left', left_on='ROADWAYSEGID', right_on='ROADWAYSEGID')

print('Number of entries in 311 data that did not map to a STREETSEGID in the MAR addresses table: ', pd.isnull(data_311_merged['STREETSEGID']).sum())
print('Percent of entries in 311 data that did not map to a STREETSEGID in the MAR addresses table: ', \
      pd.isnull(data_311_merged['STREETSEGID']).sum()/len(data_311_merged['STREETSEGID'])*100)
```
