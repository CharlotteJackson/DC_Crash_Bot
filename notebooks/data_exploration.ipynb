```python
#geopandas.datasets.available
#dont like geopands will use folium


import folium
```

```python
def mark_map(row, m, map_lat, map_long, cols):

    curr_lat = row[map_lat]
    curr_long = row[map_long]

    tooltip_string = ""

    for col in cols:
        tooltip_string += "<b>" + str(col) + ": </b>" + str(row[col]) + "<br>"
        
    
    try:

        folium.Marker(
            [curr_lat, curr_long], popup=tooltip_string, tooltip=tooltip_string
        ).add_to(m)
    except:
        return None

    
#use sample data
#crash_df = pd.read_csv("data/crashes/Crashes_in_DC.csv").head(25)

#vision_three_merge
    
map_lat = "Y"
map_long = "X"


first_lat = crash_df[map_lat].mean()
first_long = crash_df[map_long].mean()

# fill nans
#crash_df = crash_df.fillna("NULL")

crash_three_merge = crash_three_merge.fillna("NULL")

m = folium.Map(location=[first_lat, first_long], zoom_start=13)

# now place a marker for each row in df
crash_three_merge.apply(lambda row: mark_map(row, m, map_lat, map_long, crash_three_merge.columns), axis=1)    
    
crash_three_merge
```
